```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
import time

ser = Service('chromedriver.exe')
driver = webdriver.Chrome(service = ser)

url = 'https://www.instagram.com/'
driver.get(url)
```

### 인스타그램 창이 뜨면 로그인 진행 후 아래의 함수들을 실행시켜주세요!

```python


def insta_searching(word):
    url = 'https://www.instagram.com/explore/tags/{}/'.format(word)
    return(url)

from selenium.webdriver.common.by import By

def select_first(driver):
    # 웹페이지에 접속해서 첫번째 게시글 해당하는 tag 선택해줘
    first = driver.find_element(By.CSS_SELECTOR, 'div._aagu')

    first.click()
    time.sleep(3)

import re
def get_content(driver):
    html_insta = driver.page_source
    soup = BeautifulSoup(html_insta, 'html.parser')

    try:
        content = soup.select('div._a9zs > span ')[0].text
        print(content)

    except:
        content = " "

    try:
        tags = re.findall(r'#[^\s#,\\]+',content)
    except:
        tags = " "

    data = [content, tags]
    print(data ,"!!!")
    return(data)

def move_next(driver):
    right = driver.find_element(By.CSS_SELECTOR,"div._aaqg")

    right.click()
    time.sleep(3)

def insta_crawling(word, n):

    url = insta_searching(word)
    driver.get(url)
    time.sleep(8)


    # 첫 사진 선택
    select_first(driver)
    time.sleep(8)


    results = []
    for i in range(n):
        time.sleep(2)
        try:
            data = get_content(driver)
            results.append(data)
            move_next(driver)

        except:
            time.sleep(2)
            move_next(driver)

    return(results)
```

### tag 변수에 크롤링 할 태그 입력

```python
tag = "공유주택"
results = insta_crawling(tag, 100)
results_df = pd.DataFrame(results)
results_df.columns = ['content','tags']
```

### 크롤링한 파일에서 단어 빈도수 확인

```python
def word_count_cal(filename,remove_list):
    text = ""
    with open(filename,"r",encoding='utf-8') as f:
        lines = f.readlines()
        for i in lines :
            text += i
            
    for n in remove_list:
        text = text.replace(n,'',-1)
        
    word_count = WordCloud().process_text(text)

    one_count_arr = []
    for w in word_count.items():
        if w[1] < 3:
            one_count_arr.append(w[0])
            
    for o in one_count_arr:
        del word_count[o]
        
    sorted_count = sorted(word_count.items(), key=lambda x:x[1],reverse=True)
    sum_count = sum(word_count.values())
    return sorted_count, sum_count
```

- 검색태그와 일치하거나 매우 유사한 경우 제거
- 인스타에서 광고 혹은 일상적으로 쓰이는 단어 제거

```python
remove_list = ['맥그리들','맥모닝','맥도날드','mcdonald','세트','s','맥날','햄버거',\
               '먹스타그램','맛스타그램','좋아요','선팔']
sorted_count_dict, sum_count = word_count_cal("./data/맥모닝_202210231544_tag.txt",remove_list)
sorted_count_dict
```
